```
* N숫자/./_ -> N
* AD숫자/./_ -> AD
* T숫자/./_ -> T
* GSM숫자/./_ -> GSM

* 알파벳+숫자/_ -> 알파벳
* 알파벳+. -> 1) 알파벳+.+알파벳 / 2) 알파벳+.+숫자 -> 알파벳
```

````python
import re
alphaNumORDot = re.compile('[a-zA-Z]+[0-9]+|[a-zA-Z]+_')
alphaDotAlpha = re.compile('[a-zA-Z]+\.[a-zA-Z]+')
alphaDotNum = re.compile('[a-zA-Z]+\.[0-9]+')
onlyAlpha = re.compile('[a-zA-Z]+')

rowNameSet = set()

for i in list(predict.index):
    if alphaNumORDot.match(i): # 앞에 알파벳 부분만 추출  
        matchStr = onlyAlpha.match(i)
        
    elif alphaDotAlpha.match(i):
        matchStr = alphaDotAlpha.match(i)
        
    elif alphaDotNum.match(i): # 앞에 알파벳 부분만 추출  
        matchStr = onlyAlpha.match(i)
        
    else:
        print(i)
        matchStr = i
        
    rowNameSet.add(matchStr.group())

rowNameList = list(rowNameSet)
print(rowNameList)
````

````python
# pearson correlation 구하는 함수 
def corr_pearson(df, row_name):
    data_row = df[df.columns.difference(['rowName'])].T

    df_corr_row_pearson = data_row.corr(method='pearson')
    csvName = "corr_pearson_rowName_"+row_name +".csv"
    df_corr_row_pearson.to_csv(csvName) # csv 파일로 correlation 값 저장 
    return df_corr_row_pearson
````

````python
# spearman correlation 구하는 함수 
def corr_spearman(df, row_name):
    data_row = df[df.columns.difference(['rowName'])].T

    df_corr_row_spearman = data_row.corr(method='spearman')
    csvName = "corr_spearman_rowName_"+row_name +".csv"
    df_corr_row_spearman.to_csv(csvName) # csv 파일로 correlation 값 저장 
    return df_corr_row_spearman
````

````python
# pearson heatmap 구하는 함수
def heatmap_pearson(df, row_name):

    import matplotlib.pyplot as plt 
    import seaborn as sns   

    plt.figure(figsize=(15,15))
    ax = sns.heatmap(data = df, linewidths=.5, annot=False, cmap="Spectral") # , cmap='Reds'
    plt.title('Heatmap of pearson pandas correlation for rowName '+row_name, fontsize=20)

    ax.set_xticklabels(labels = df.columns, rotation=45)

    imagename="heatmap_pearson_rowName_"+row_name+".png"
    plt.savefig(imagename)
````

````python
# spearman heatmap 구하는 함수
def heatmap_spearman(df, row_name):

    import matplotlib.pyplot as plt 
    import seaborn as sns   

    plt.figure(figsize=(15,15))
    ax = sns.heatmap(data = df, linewidths=.5, annot=False, cmap="Spectral") # , cmap='Reds'
    plt.title('Heatmap of spearman pandas correlation for rowName '+row_name, fontsize=20)

    ax.set_xticklabels(labels = df.columns, rotation=45)

    imagename="heatmap_spearman_rowName_"+row_name+".png"
    plt.savefig(imagename)
````
